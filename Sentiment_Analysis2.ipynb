```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(num_words=20000)
tokenizer.fit_on_texts(new_df["text"])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)+1
```

```python
texts =["What are you doing","Ayanokoji is genius boy"]
sequences = tokenizer.texts_to_sequences(texts)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(new_df["text"],new_df["target"],test_size=0.2,random_state=42)
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)
```

```python
X_train = tf.keras.utils.pad_sequences(
    X_train,
    maxlen=50,
    dtype='int32',
    padding='post',
    truncating='post',
    value=0.0
)
X_test = tf.keras.utils.pad_sequences(
    X_test,
    maxlen=50,
    dtype='int32',
    padding='post',
    truncating='post',
    value=0.0
)
```

```python
X_train
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
Y_train = np.array(Y_train)
Y_test = np.array(Y_test)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(input_dim=20000, output_dim=128, input_length=50))
model.add(tf.keras.layers.LSTM(128))
model.add(tf.keras.layers.Dense(1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

model.summary()


early_stop = EarlyStopping(
    monitor='val_loss',
    patience=2,
    restore_best_weights=True
)

history = model.fit(
    X_train, Y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=64,
    callbacks=[early_stop]
)
```

```python
# @title Default title text
# Get probabilities
y_probs = model.predict(X_test)   # shape (num_samples, 1)

# Convert to labels (0 or 1)
y_preds = (y_probs >= 0.5).astype(int).flatten()  # shape (num_samples,)

# Compute accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_test, y_preds)
print("Test Accuracy:", acc)
```
